# BUILDING THE DECISION MODEL

```r
install.packages("rpart.plot")
```

```r
install.packages("caret")
```

```r
#LOADING THE REQUIRED PACKAGES
library(rpart)
library(rpart.plot)
library(caret)
```

```r
#DROPPING THE UNNECESSARY VARIABLES
df1 <- df[-c(1)]
head(df1)
```

```r
#SPLITTING THE DATA INTO TRAINING AND TESTING DATA
set.seed(1234)
train_sample <- sample(1:nrow(df1),0.7*nrow(df1))
train <- df1[train_sample,]
test <- df1[-train_sample,]
```

```r
#BUILDING THE DECISION TREE MODEL
dt_model <- rpart(booking_status ~ .,data=train)
```

```r
#PLOTTING THE MODEL
rpart.plot(dt_model)
```

```r
#MAKING THE PREDICTIONS
dt_prediction <- predict(dt_model,test)
head(dt_prediction,20)
```

```r
# 1. Interpret the Model
# Print the decision tree model summary
print(dt_model)

# 2. Variable Importance
# Print the variable importance table
print(dt_model$variable.importance)

# 3. Make predictions on the test data
# Extract the predicted class probabilities for each observation
dt_prediction <- predict(dt_model, test, type = "class")

# 4. Evaluate the model's performance
# Create a confusion matrix
conf_matrix <- table(Actual = test$booking_status, Predicted = dt_prediction)
print(conf_matrix)

# Calculate accuracy
accuracy <- sum(diag(conf_matrix)) / sum(conf_matrix)
print(paste("Accuracy:", accuracy))

# Calculate error rate
error_rate <- 1 - accuracy
print(paste("Error Rate:", error_rate))

# Calculate other relevant metrics as needed
```

# BUILDING THE LOGISTIC REGRESSION MODEL

```r
install.packages("car")
```

```r
#LOADING THE REQUIRED PACKAGES
library(car)
library(caret)
```

```r
#BUILDING THE MODEL
glm_model <- glm(booking_status~lead_time+market_segment_type,data=df1,family=binomial)

summary(glm_model)
```

```r
#CHECKING FOR MULTICOLLINEARITY
vif(glm_model)
```

```r
#PREDICT FOR THE MODEL
glm_prediction <- predict(glm_model,test)
head(glm_prediction,20)
```

```r
# 1. Interpret the Model
# Print the summary of the logistic regression model
print(summary(glm_model))

# 2. Checking for Multicollinearity
# Calculate Variance Inflation Factors (VIF)
vif_values <- car::vif(glm_model)
print(vif_values)

# 3. Make predictions on the test data
# Predict the probabilities of booking_status being "Canceled" or "Not_Canceled"
glm_prediction <- predict(glm_model, test, type = "response")

# Convert predicted probabilities to class predictions
glm_prediction_class <- ifelse(glm_prediction > 0.5, "Not_Canceled", "Canceled")

# Compare actual vs. predicted class labels
conf_matrix <- table(Actual = test$booking_status, Predicted = glm_prediction_class)
print(conf_matrix)

# Calculate accuracy
accuracy <- sum(diag(conf_matrix)) / sum(conf_matrix)
print(paste("Accuracy:", accuracy))

# Calculate error rate
error_rate <- 1 - accuracy
print(paste("Error Rate:", error_rate))

# Calculate other relevant metrics as needed
```

# FORECASTING

```r
install.packages("forecast")
```

```r
#LOADING THE REQUIRED PACKAGES
library(forecast)

# SPLITTING THE DATA INTO TRAINING AND TESTING DATA
set.seed(1234)
train_indices <- sample(1:nrow(df1), 0.7 * nrow(df1))
train <- df1[train_indices, ]
test <- df1[-train_indices, ]

# BUILDING THE ARIMA MODEL
arima_model <- auto.arima(train$avg_price_per_room)

# FORECASTING WITH ARIMA
arima_forecast <- forecast(arima_model, h = nrow(test))

# BUILDING THE ETS MODEL
ets_model <- ets(train$avg_price_per_room)

# FORECASTING WITH ETS
ets_forecast <- forecast(ets_model, h = nrow(test))

# PLOTTING ARIMA FORECAST
plot(arima_forecast, main = "ARIMA Forecast", xlab = "Time", ylab = "Average Price")

# PLOTTING ETS FORECAST
plot(ets_forecast, main = "ETS Forecast", xlab = "Time", ylab = "Average Price")
```
